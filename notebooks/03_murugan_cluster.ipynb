# Proof 3 — Murugan Semantic Cluster

**Hypothesis:** Signs P316 (வேல் vel — spear), P122 (மீன் meen — fish/star), and P062 (மலை malai — mountain) form a non-random semantic cluster representing the three primary attributes of the Tamil deity **Murugan**:

- வேல் = Murugan's spear (his primary weapon and symbol)
- மீன் = star (Murugan's celestial identity; fish/star homophone in Tamil)
- மலை = mountain (Murugan is the god of hills — *குறிஞ்சி*)

If these three signs appear together more often than random chance predicts, they form a **semantic field** — evidence of religious/titular vocabulary.

**Method:** Chi-square test of co-occurrence against expected frequency under independence.

```python
import sys
sys.path.insert(0, '..')
from indus_decode import CORPUS, SIGN_MAP, proof3_murugan_cluster
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import seaborn as sns
from collections import Counter
```

## 1. The Murugan Sign Triad

```python
cluster = ('P316', 'P122', 'P062')

print('MURUGAN ATTRIBUTE SIGNS')
print('=' * 45)
for sign in cluster:
    info = SIGN_MAP[sign]
    print(f"  {sign}  {info['tamil']:10s}  {info['roman']:8s}  → {info['meaning']}")
    print(f"         DEDR: {info['dedr']}")
```

## 2. Chi-Square Co-occurrence Test

```python
result = proof3_murugan_cluster()

print('PROOF 3 — MURUGAN CLUSTER RESULTS')
print('=' * 45)
for k, v in result.items():
    if k != 'cluster_signs':
        print(f'  {k:25s}: {v}')
```

## 3. Sign Co-occurrence Heatmap (Top 10 Signs)

```python
sign_counts = Counter(s for seal in CORPUS for s in seal['seq'])
top10 = [s for s, _ in sign_counts.most_common(10)]
labels = [SIGN_MAP.get(s, {}).get('roman', s) for s in top10]

# Build co-occurrence matrix
matrix = np.zeros((10, 10))
for seal in CORPUS:
    for i, s1 in enumerate(top10):
        for j, s2 in enumerate(top10):
            if s1 in seal['seq'] and s2 in seal['seq'] and i != j:
                matrix[i][j] += 1

# Highlight Murugan cluster
murugan_idx = [top10.index(s) for s in cluster if s in top10]

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(matrix, xticklabels=labels, yticklabels=labels,
            cmap='YlOrRd', annot=True, fmt='.0f', ax=ax,
            linewidths=0.5, linecolor='white')

# Draw box around Murugan cluster
if len(murugan_idx) >= 2:
    min_idx, max_idx = min(murugan_idx), max(murugan_idx)
    rect = plt.Rectangle((min_idx, min_idx), max_idx - min_idx + 1, max_idx - min_idx + 1,
                          fill=False, edgecolor='#2980B9', linewidth=3)
    ax.add_patch(rect)

ax.set_title('Sign Co-occurrence Matrix (Top 10 Signs)\nBlue box = Murugan semantic cluster', fontsize=13)
plt.tight_layout()
plt.savefig('proof3_cooccurrence.png', dpi=150, bbox_inches='tight')
plt.show()
```
