# Proof 4 — Fish→Jar Agglutination Ratio

**Hypothesis:** The bigram P122→P385 (*meen* + *-an* = *மீனன்* Meenan, 'The Starry One') appears far more often than random chance predicts, demonstrating **agglutinative word formation** — the core grammatical feature of Tamil.

Agglutination means words are built by gluing a root to suffixes:
- Root: மீன் (meen) — fish/star
- Suffix: -அன் (-an) — masculine nominal suffix  
- Result: மீனன் (Meenan) — a divine title meaning 'The Starry One' or 'Lord of Stars'

**Method:** Observed bigram count vs. expected under sign independence. Ratio > 1 = agglutination signal.

```python
import sys
sys.path.insert(0, '..')
from indus_decode import CORPUS, SIGN_MAP, proof4_agglutination
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

## 1. Bigram Analysis: Observed vs Expected

```python
result = proof4_agglutination()

print('PROOF 4 — AGGLUTINATION RATIO')
print('=' * 45)
for k, v in result.items():
    print(f'  {k:25s}: {v}')

print(f'\nP122→P385 appears {result["agglutination_ratio"]}x more than chance predicts.')
print(f'Tamil reading: {result["reading"]}')
```

## 2. Top Bigrams — Observed vs Expected

```python
sign_counts = Counter(s for seal in CORPUS for s in seal['seq'])
total = sum(sign_counts.values())
n_seals = len(CORPUS)

bigram_obs = Counter()
for seal in CORPUS:
    for i in range(len(seal['seq']) - 1):
        bigram_obs[(seal['seq'][i], seal['seq'][i+1])] += 1

ratios = []
for (a, b), obs in bigram_obs.most_common(30):
    exp = n_seals * (sign_counts[a] / total) * (sign_counts[b] / total)
    if exp > 0.5:
        ratios.append(((a, b), obs, round(exp, 2), round(obs / exp, 2)))

ratios.sort(key=lambda x: x[3], reverse=True)
top = ratios[:12]

labels  = [f"{SIGN_MAP.get(a,{}).get('roman',a)}→{SIGN_MAP.get(b,{}).get('roman',b)}" for (a,b),_,_,_ in top]
r_vals  = [r for _,_,_,r in top]
colors  = ['#C0392B' if labels[i].startswith('meen→-an') else '#2980B9' for i in range(len(labels))]

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(labels, r_vals, color=colors, edgecolor='white')
ax.axhline(y=1, color='black', linestyle='--', alpha=0.5, label='Random baseline (ratio=1)')
ax.set_title('Bigram Agglutination Ratios (Observed / Expected)\nRed = மீன்→-அன் (meen→-an)', fontsize=13)
ax.set_ylabel('Ratio (observed / expected)')
ax.set_xlabel('Bigram')
plt.xticks(rotation=45, ha='right')
ax.legend()
plt.tight_layout()
plt.savefig('proof4_bigrams.png', dpi=150, bbox_inches='tight')
plt.show()
```
